# OrdinalEncoder
- An ordinal encoding involves mapping each unique label to an integer value. This type of encoding is really only appropriate if there is a known relationship between the categories.
- OrdinalEncoder is for converting features, while LabelEncoder is for converting target variable.

```python
from sklearn.preprocessing import OrdinalEncoder
df1.dropna(inplace=True)
df1.isna().sum()
```

```python
df1.isnull().sum()
```

```python
OE=OrdinalEncoder()
df1["sex_oe"]=OE.fit_transform(df1[["sex"]])   # Double Square bracket if 2-D error occurs
df1.head()
```
